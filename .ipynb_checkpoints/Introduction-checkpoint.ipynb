# Introduction
**Dataset Source**: https://archive.ics.uci.edu/ml/datasets/Cervical+cancer+%28Risk+Factors%29

I will be analyzing data on cervical cancer risk factors. The researchers surveyed almost 900 women at the the University Hospital of Caracas in Caracas, Venezuela. The dataset includes numerical information such as the age of participants, number of sexual partners, or years using hormonal contraceptives and boolean information such as does the participant smoke, have HPV, or use an IUD. 

I wanted to investigate this topic because as a woman, women's health is important to me. Women with health concerns are often discredited and their concerns ignored. It is all to common to hear stories of women going to their doctor for help and being told they're wrong or overreacting, only to find out later their condition was as serious as they had thought and the doctor was wrong. It is important for women to be aware of behaviors that could be putting them at risk to protect themselves or to be able to recognize when their symptoms could be part of a bigger issue. The sooner a health condition is caught, especially one as serious as cancer, the sooner treatment can begin, and the better odds someone has of recovering. 

In this project, I want to identify the behaviors most closely associated with getting cervical cancer and if there are behaviors when combined, make the risk higher. 

# Methods

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/gbraga12/DH140-FinalProject/main/cervical_cancer_risk_factors.csv')
```

To begin my analysis, I had to edit the data from the csv file. The author had used '?' to denote information they did have. These values are not simple to work with so I needed to change them to None values, which is understood as a null value in Python. I also needed to change all the datatypes to integers. The authors had originally denoted some values as boolean but that did not transfer over to my Jupyter notebook in a way that I could easily use. By turning all the values into integers, it was simple to use the data as I wished

```python
# Change ? in dataset to None
for i in data:
    data.loc[data[i] == '?', i] = None

# Convert at column datatypes into int
columns = data.columns
for j in columns:
    data[j] = pd.to_numeric(data[j])
```

```python
G = nx.Graph()

G.add_node('condylomatosis')
G.add_node('cervical condylomatosis')
G.add_node('vaginal condylomatosis')
G.add_node('vulvo-perineal condylomatosis')
G.add_node('syphilis')
G.add_node('pelvic inflammatory disease')
G.add_node('genital herpes')
G.add_node('molluscum contagiosum')
G.add_node('AIDS')
G.add_node('HIV')
G.add_node('Hepatitis B')
G.add_node('HPV')

colors = []

for i in range(data.shape[0]):
    if data.loc[i,'STDs:condylomatosis']:
        G.add_node(i)
        G.add_edge(i, 'condylomatosis')
        colors.append('red')
    if data.loc[i,'STDs:cervical condylomatosis']:
        G.add_node(i)
        G.add_edge(i, 'cervical condylomatosis')
        colors.append('orange')
    if data.loc[i,'STDs:vaginal condylomatosis']:
        G.add_node(i)
        G.add_edge(i, 'vaginal condylomatosis')
        colors.append('yellow')
    if data.loc[i,'STDs:vulvo-perineal condylomatosis']:
        G.add_node(i)
        G.add_edge(i, 'vulvo-perineal condylomatosis')
        colors.append('green')
    if data.loc[i,'STDs:syphilis']:
        G.add_node(i)
        G.add_edge(i, 'syphilis')
        colors.append('blue')
    if data.loc[i,'STDs:pelvic inflammatory disease']:
        G.add_node(i)
        G.add_edge(i, 'pelvic inflammatory disease')
        colors.append('purple')
    if data.loc[i,'STDs:genital herpes']:
        G.add_node(i)
        G.add_edge(i, 'genital herpes')
        colors.append('pink')
    if data.loc[i,'STDs:molluscum contagiosum']:
        G.add_node(i)
        G.add_edge(i, 'molluscum contagiosum')
        colors.append('black')
    if data.loc[i,'STDs:AIDS']:
        G.add_node(i)
        G.add_edge(i, 'AIDS')
        colors.append('brown')
    if data.loc[i,'STDs:HIV']:
        G.add_node(i)
        G.add_edge(i, 'HIV')
        colors.append('gray')
    if data.loc[i,'STDs:Hepatitis B']:
        G.add_node(i)
        G.add_edge(i, 'Hepatitis B')
        colors.append('lavender')
    if data.loc[i,'STDs:HPV']:
        G.add_node(i)
        G.add_edge(i, 'HPV')
        colors.append('teal')
```
